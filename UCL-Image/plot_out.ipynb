```python
import matplotlib
#matplotlib.use("AGG")
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def plot(out_list):
    # process name
    fname_list = []
    for fname in out_list:
        fname_list.append(os.path.join(os.path.join("./out", fname),"te_acc.npy"))
    
    # open result and plot
    for i, fname in enumerate(fname_list):
        acc_list = np.load(fname)
        plt.plot(acc_list, label=out_list[i])
        
    # plt.ylim([0.55,0.63])
    plt.legend()
    plt.show()
```

```python
out_list = ["dclif_small_cifar10_DEF","nocl_small_cifar10_DEF"]
```

```python
plot(out_list)
```
