```python
import matplotlib
#matplotlib.use("AGG")
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
def plot(out_list):
    # process name
    fname_list = []
    for fname in out_list:
        fname_list.append(os.path.join(os.path.join("./out", fname),"te_acc.npy"))
    
    # open result and plot
    for i, fname in enumerate(fname_list):
        acc_list = np.load(fname)
        plt.plot(acc_list, label=out_list[i])
        
    # plt.ylim([0.55,0.63])
    plt.legend()
    plt.show()
```

```python
# plot uncertainty score distributions
data_name = "cifar10"
mode = "epis"
score_ucl_path = "./out/ucl_small_{}_DEF/score_{}.npy".format(data_name, mode)
score_ucltl_path = "./out/bcltl_small_{}_DEF/score_{}.npy".format(data_name, mode)

score_ucl = np.load(score_ucl_path)
score_ucltl = np.load(score_ucltl_path)

# standardize
# func = lambda x: (x-x.min())/(x.max()-x.min())
func = lambda x:x
fig, ax = plt.subplots()
sns.distplot(func(score_ucl), kde=False, bins=20, label="self-tutoring")
sns.distplot(func(score_ucltl), kde=False, bins=20,label="transfer")
# ax.set_ylim([0,20])
plt.legend()
plt.show()
```
